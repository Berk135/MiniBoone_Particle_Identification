<h2> 2. Exploratory Data Analysis </h2>

Now, I will explore the MiniBooNE data set.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
#Get data from the parquet file
df = pd.read_parquet('data/miniboone.parquet')
```

```python
#Get the shape of the data & take a look at first 5 rows
print(df.shape)
df.head()
```

The data set has 130064 rows and 51 columns, the last one being the particle (target) whereas the rest are the 50 particle ID columns (features).

From the [data set information](https://archive.ics.uci.edu/ml/datasets/MiniBooNE+particle+identification) and the first notebook, we know that there are 36499 electron neutrinos (designated as 1's) and 93565 muon neutrinos (designated as 0's).

```python
print("Number of electron neutrinos: ", len(df[df["particle"]==1]))
print("Number of muon neutrinos: ",len(df[df["particle"]==0]))
```

<h3>2.1 Nullity and Duplicate Checks</h3>

```python
#check for missing values
df.isnull().sum()
```

There are no missing values for any of the columns. Now, I'll check for duplicate rows in the data set.

```python
#check duplicate rows
df[df.duplicated()]
```

```python
#check duplicate rows for electron neutrinos
print("Number of duplicate rows that are electron neutrinos: ", len(df[df.duplicated() & (df["particle"]==1)]))
```

```python
#check duplicate rows for muon neutrinos
print("Number of duplicate rows that are muon neutrinos: ", len(df[df.duplicated() & (df["particle"]==0)]))
```

Looking at the duplicate entries, we see that there are 466 duplicate entries of which most are muon neutrinos. Since we see these entries are skewed heavily towards muon neutrinos, removing these duplicates can affect the generalization performance of the eventual model in the end. Thus, I will keep these duplicates.

<h3>2.3 Data Visualization</h3>

Now, I will check the distribution of values in each column.

```python
plt.figure(figsize=(8,6))
plt.style.use("seaborn")
value_dist = df["particle"].value_counts()
plt.pie(x=value_dist, autopct="%.1f%%", labels = value_dist.keys(), textprops={'fontsize': 14})
```

We see that electron neutrinos (encoded as 1's in the data set) are about 28% of the data while the rest are muon neutrinos (encoded as 0's in the data set).

Now, for the distribution for each column:

```python
# plot histograms for each column
fig = plt.figure(figsize=(16,16))
plt.style.use("seaborn")
cols = df.drop("particle", axis=1).columns

i=1
for c in cols:
    ax = fig.add_subplot(13, 4, i)
    sns.distplot(df[c], kde = False, color="#000080")
    i+=1

fig.tight_layout()
plt.show()
```

Every column contains -999 as the lower bound value which skews the distribution plots and restricts the visibility of the variability of other values in a given column. Thus, let's remove these -999's from columns to get better visuals

```python
df["C2"][df["C2"]==-999]
```

```python
def remove_minus_999(col):
    copy = col.copy()
    copy = copy[copy!=-999]

    return copy
```

```python
#get distribution plots again after removing -999
fig = plt.figure(figsize=(16,16))
plt.style.use("seaborn")
cols = df.drop("particle", axis=1).columns

i=1
for c in cols:
    ax = fig.add_subplot(13, 4, i)
    sns.distplot(remove_minus_999(df[c]), kde = False, color="#000080")
    i+=1

fig.tight_layout()
plt.show()
```

We now see the structure of the distributions better. They seem to have different distributions from each other. For example, column C34 shows a prominent left skew whereas C41 seems more normally distributed.

<h3>2.4 Correlations</h3>

```python
df.drop("particle", axis=1).corr()
```

We see from the correlation matrix that many columns are highly correlated. This is likely due to the fact that most columns have the extreme value of -999. We can remove these from the data set to have a better analysis.

```python
#create the filtered df3
df = df[df["C1"]!=-999]
df3 = df.drop("particle", axis=1)
```

```python
df3.describe()
```

Having removed the extreme values, we can look at the correlations once again. Heatmaps can be used for this purpose but since we have 50 features, it's best if they are split into 2 batches, each of 25 features to see their correlations.

```python
#create batches of columns
col_batch1 = ["C" + str(i+1) for i in range(25)]
col_batch2 = ["C" + str(i+1) for i in range(25, 50)]
```

```python
#check correlations among first 25 features using a clustermap (heatmap + dendrogram)
plt.figure(figsize=(1,1))
sns.clustermap(round(df3.corr()[col_batch1].drop(col_batch2),2), annot=True, center = 0, cmap="vlag")
```

```python
#check correlations between first 25 features and the last 25 using a clustermap (heatmap + dendrogram)
plt.figure(figsize=(1,1))
sns.clustermap(round(df3.corr()[col_batch1].drop(col_batch1),2), annot=True, center = 0, cmap="vlag")
```

```python
#check correlations among second 25 features using a clustermap (heatmap + dendrogram)
plt.figure(figsize=(1,1))
sns.clustermap(round(df3.corr()[col_batch2].drop(col_batch1),2), annot=True, center = 0, cmap="vlag")
```

As can be seen in the correlation matrices, we don't have near-perfect collinearities once we remove the extreme values. 
We see the variety of correlations from negative to positives and even near zero correlation between column C20 and other columns.
